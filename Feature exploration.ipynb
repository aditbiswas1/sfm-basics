```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image1 = cv2.imread('test_images/P1180141.JPG')
```

```python
plt.imshow(image1)
```

```python
sift = cv2.SIFT()
```

```python
gray= cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)
k, d = sift.detectAndCompute(gray, None)
```

```python
imgsift =cv2.drawKeypoints(gray,k)
```

```python
plt.imshow(imgsift)
```

```python
orb = cv2.ORB(40000)
```

```python
kp = orb.detect(image1)
```

```python
img_orb = cv2.drawKeypoints(image1, kp)
```

```python
plt.imshow(img_orb)
```
